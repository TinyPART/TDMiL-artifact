```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
iid_dfs = [
    pd.read_csv(
        f"train_iid_cancer_norm_{idx}.csv",
        header=None,
    )
    for idx in range(5)
]
niid_dfs = [
    pd.read_csv(
        f"train_niid_cancer_norm_{idx}.csv",
        header=None,
    )
    for idx in range(5)
]
```

```python
iid_dfs[0].head()
```

```python
for idx, iid_df in enumerate(iid_dfs):
    iid_df.rename(columns={30: 'labels'}, inplace=True)
for idx, niid_df in enumerate(niid_dfs):
    niid_df.rename(columns={30: 'labels'}, inplace=True)
```
